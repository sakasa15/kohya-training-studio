# 🎨 Kohya_ss Model Downloader

**Easy model download for Kohya_ss training**  
学習に使用するモデルを簡単にダウンロードできます

## 📦 How to Use / 使い方
1. Select a cell and press **Shift + Enter** to run  
   セルを選択して **Shift + Enter** で実行
2. Or click the ▶ button at the top  
   または上部の ▶ ボタンをクリック

---

```python
# Setup / セットアップ
import os
import urllib.request

# Download model_downloader.py if not exists
if not os.path.exists('model_downloader.py'):
    print('Downloading model_downloader.py...')
    urllib.request.urlretrieve(
        'https://raw.githubusercontent.com/sakasa15/kohya-training-studio/main/scripts/model_downloader.py',
        'model_downloader.py'
    )
    print('✅ Done!')
else:
    print('✅ model_downloader.py already exists')

# Install dependencies
!pip install requests tqdm -q
```

## 🔑 Hugging Face Token設定 (Flux・SD3.5を使う場合)

FluxやSD3.5をダウンロードする場合は、以下のセルにHugging FaceのAccess Tokenを入力して実行してください。  
SD1.5・SDXLのみ使用する場合はスキップできます。

**トークンの取得方法:** https://huggingface.co/settings/tokens

```python
import os

# ここにHugging FaceのAccess Tokenを貼り付けてください
# Paste your Hugging Face Access Token here
HF_TOKEN = ""  # ← 例: hf_xxxxxxxxxxxxxxxxxxxxxxxx を""の間に入力してください

if HF_TOKEN:
    os.environ["HF_TOKEN"] = HF_TOKEN
    print("✅ HF Token set successfully / HFトークンを設定しました")
    print("   Flux and SD3.5 downloads are now enabled")
else:
    print("⚠️  No token set - Flux/SD3.5 will not be available")
    print("   SD1.5/SDXL/Anime models are still available")
```
